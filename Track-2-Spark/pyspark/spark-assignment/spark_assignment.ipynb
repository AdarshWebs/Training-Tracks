```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lower, regexp_replace

spark = (
    SparkSession.builder
    .appName("SparkJupyterAssignment")
    .master("local[*]")
    .getOrCreate()
)

print("Spark Started")
```

```python
df = spark.read.csv(
    "data/startup.csv",
    header=True,
    inferSchema=True
)

df.show(5)
```

```python
df_clean = df.withColumn(
    "startup_name_processed",
    regexp_replace(col("Startup_Name"), " ", "")
)

df_clean = df_clean.withColumn(
    "City",
    lower(col("City"))
)

df_clean.show(5)
```

```python
df_clean.printSchema()
```

```python
df_clean.write.format("jdbc") \
.option("url", "jdbc:mysql://localhost:3306/startup_db") \
.option("driver", "com.mysql.cj.jdbc.Driver") \
.option("dbtable", "startup_data_jupyter") \
.option("user", "root") \
.option("password", "adarsh02022004") \
.mode("overwrite") \
.save()
```
